```python
import pandas as pd
import numpy as np
```

```python
csv_root = 'D:\dev_folder\\data\\Nose\\MIrint\\dust.csv'
temp_dust_data = pd.read_csv(csv_root)
temp_dust_data = temp_dust_data[['시도', '측정일시', 'SO2', 'O3', 'NO2', 'PM10', 'PM25']]
temp_dust_data.head()
```

```python
temp_dust_data['측정일시'] = temp_dust_data['측정일시'].div(100).round(0).astype(int)
temp_dust_data.head()
```

```python
temp_dust_data_avg = temp_dust_data.groupby(['시도','측정일시']).mean()
temp_dust_data_re = temp_dust_data_avg.reset_index()
temp_dust_data_re.head()
```

```python
save_name = 'dust2.csv'
temp_dust_data_re.to_csv(save_name, mode = 'w')
```
