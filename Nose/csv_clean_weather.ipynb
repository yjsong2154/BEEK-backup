```python
import pandas as pd
import numpy as np
```

```python
CSV_ROOT = 'C:\dev_folder\dip\TensorFlow\Data voucher\Mirint data\data\weather\날씨 일별 정보_2014~2018.csv'

data_weather = pd.read_csv(CSV_ROOT)
data_weather = data_weather[['위치', '시간', '평균 기온', '최고 기온', '최저 기온',  '강수량', '풍속', '습도']]

data_weather.head()
```

```python
print(data_weather['위치'].unique())
```

```python
loc_list = {
    '강릉' : '강원'
    , '속초' : '강원'
    , '북춘천' : '강원'
    , '동두천' : '경기'
    , '수원' : '경기'
    , '파주' : '경기'
    , '거창' : '경남'
    , '의령' : '경남'
    , '진주' : '경남'
    , '경주' : '경북'
    , '구미' : '경북'
    , '영주' : '경북'
    , '순천' : '전남'
    , '목포' : '전남'
    , '보성군' : '전남'
    , '서귀포' : '제주'
    , '고창군' : '전북'
    , '군산' : '전북'
    , '전주' : '전북'
    , ' 금산' : '충남'
    , '보령' : '충남'
    , '부여' : '충남'
    , '보은' : '충북'
    , '제천' : '충북'
    , '청주' : '충북'
}
sido_list = ['강원','경기','경남','경북','광주','대구','대전','부산','서울','세종','울산','인천','전남','전북','제주','충남','충북']

data_weather['위치'] = data_weather['위치'].replace(loc_list)
data_weather_sido = data_weather[data_weather['위치'].isin(sido_list)]
data_weather_sido.head()
```

```python
data_weather_avg = data_weather_sido.groupby(['위치','시간']).mean()
data_weather_avg.head()
save_name = 'weather.csv'
data_weather_avg.to_csv(save_name, mode = 'w')
```
