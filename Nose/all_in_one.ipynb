데이터 분석

```python
import pandas as pd
import numpy as np
```

```python
sido_list = ['강원','경기','경남','경북','광주','대구','대전','부산','서울','울산','인천','전남','전북','제주','충남','충북']
```

```python
csv_root = 'disease.csv'
temp_dis_data = pd.read_csv(csv_root)
print(temp_dis_data['시도'].unique())
print(len(temp_dis_data['일자'].unique()))
temp_dis_data = temp_dis_data[['시도', '일자', '인원수']]
print(temp_dis_data.dtypes)
temp_dis_data.head()
```

```python
temp_dis_data = temp_dis_data[temp_dis_data['시도'].isin(sido_list)]
temp_dis_data['일자'] = temp_dis_data['일자'].astype(int)
temp_dis_data['day'] = temp_dis_data['일자'].astype(str)
temp_dis_data['date'] = pd.to_datetime(temp_dis_data['day'], format='%Y%m%d', errors='ignore')
temp_dis_data['년'] = temp_dis_data['일자'].div(10000).round(0).astype(int)
temp_dis_data['월'] = temp_dis_data['일자'].div(100).round(0).mod(100).astype(int)
temp_dis_data['일'] = temp_dis_data['일자'].round(0).mod(100).astype(int)
temp_dis_data = temp_dis_data.drop(['일자','day'],axis = 1)
print(len(temp_dis_data))
temp_dis_data.head()
```

```python
csv_root = 'dust.csv'
temp_dust_data = pd.read_csv(csv_root)
print(temp_dust_data['시도'].unique())
print(len(temp_dust_data['측정일시'].unique()))
print(len(temp_dust_data))
temp_dust_data = temp_dust_data[['시도', '측정일시', 'SO2', 'O3', 'NO2', 'PM10', 'PM25']]
temp_dust_data.head()
```

```python
temp_dust_data = temp_dust_data[temp_dust_data['시도'].isin(sido_list)]
temp_dust_data['측정일시'] = temp_dust_data['측정일시'].astype(int)
temp_dust_data['day'] = temp_dust_data['측정일시'].astype(str)
temp_dust_data['date'] = pd.to_datetime(temp_dust_data['day'], format='%Y%m%d', errors='ignore')
temp_dust_data['년'] = temp_dust_data['측정일시'].div(10000).round(0).astype(int)
temp_dust_data['월'] = temp_dust_data['측정일시'].div(100).round(0).mod(100).astype(int)
temp_dust_data['일'] = temp_dust_data['측정일시'].round(0).mod(100).astype(int)
temp_dust_data = temp_dust_data.drop(['측정일시','day'],axis = 1)
print(len(temp_dust_data))
temp_dust_data.head()
```

```python
csv_root = 'weather.csv'
temp_wea_data = pd.read_csv(csv_root)
print(temp_wea_data['위치'].unique())
print(len(temp_wea_data['시간'].unique()))
print(len(temp_wea_data))
temp_wea_data.head()
```

```python
temp_wea_data.rename(columns={'위치' : '시도'}, inplace = True)
temp_wea_data['date'] = pd.to_datetime(temp_wea_data['시간'], format='%Y-%m-%d', errors='ignore')
temp_wea_data['년'] = temp_wea_data.시간.str.split('-').str[0].astype(int)
temp_wea_data['월'] = temp_wea_data.시간.str.split('-').str[1].astype(int)
temp_wea_data['일'] = temp_wea_data.시간.str.split('-').str[2].astype(int)
temp_wea_data = temp_wea_data.drop(['시간'],axis = 1)
print(len(temp_wea_data))
temp_wea_data.head()
```

```python
temp_all = pd.merge(temp_dis_data, temp_dust_data)
temp_all = pd.merge(temp_all, temp_wea_data)
print(len(temp_all))
temp_all.head()
```

```python
save_name = 'all_in_one.csv'
temp_all.to_csv(save_name, mode = 'w')
```
