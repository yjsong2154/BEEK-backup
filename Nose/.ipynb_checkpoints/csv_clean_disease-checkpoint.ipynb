질병정보 처리

```python
import pandas as pd
import numpy as np
```

```python
def dis_clean(file_name):
    csv_root = 'D:\dev_folder\\data\\data\\api_data\\disease\\'+ file_name +'.csv'
    temp_dis_data = pd.read_csv(csv_root, encoding='cp949')
    temp_dis_data = temp_dis_data[['요양개시일자', '시도코드', '주상병코드']]
    
    loc_code = { 11 : '서울', 26 : '부산', 27 : '대구', 28 : '인천', 29 : '광주', 30 : '대전', 31 : '울산', 36 : '세종', 41 : '경기', 42 : '강원', 43 : '충북', 44 : '충남', 45 : '전북', 46 : '전남', 47 : '경북', 48 : '경남', 49 : '제주', 50 : '제주'}

    temp_dis_data['시도'] = temp_dis_data['시도코드'].replace(loc_code)
    temp_dis = temp_dis_data.drop(['시도코드'],axis = 1)
    
    dis_list = ['J30','J300','J301','J302','J303','J304','J31','J310']
    temp_dis = temp_dis[temp_dis['주상병코드'].isin(dis_list)]
    
    temp_dis_count = temp_dis.groupby(['시도','요양개시일자']).count()
    temp_dis_data = temp_dis_count.reset_index()
    
    return temp_dis_data
```

```python
CSV_NAMES = ['dis_2014','dis_2015','dis_2016','dis_2017']

df_temp = pd.DataFrame({'시도' : [], '요양개시일자' : [], '주상병코드' : [] })

for name in CSV_NAMES:
    print(name)
    df_temp = pd.concat([df_temp, dis_clean(name)], ignore_index=True)
              
df_temp.head()
```

```python
csv_root = 'D:\dev_folder\\data\\data\\api_data\\disease\\dis_2018.csv'
temp_dis_2018 = pd.read_csv(csv_root, encoding='cp949')
temp_dis_2018.head()
```

```python
temp_dis_2018_data = temp_dis_2018[['일자', '주소코드', '인원수']]

loc_code = { 11 : '서울', 26 : '부산', 27 : '대구', 28 : '인천', 29 : '광주', 30 : '대전', 31 : '울산', 36 : '세종', 41 : '경기', 42 : '강원', 43 : '충북', 44 : '충남', 45 : '전북', 46 : '전남', 47 : '경북', 48 : '경남', 49 : '제주', 50 : '제주'}

temp_dis_2018_data['시도'] = temp_dis_2018_data['주소코드'].replace(loc_code)
temp_dis = temp_dis_2018_data.drop(['주소코드'],axis = 1)

temp_dis_count = temp_dis.groupby(['시도','일자']).sum()
temp_dis_data = temp_dis_count.reset_index()

temp_dis_data['인원수'] = temp_dis_data['인원수'].div(56)

temp_dis_data.head()
```

```python
df_temp.rename(columns={"요양개시일자": "일자", "주상병코드": "인원수"}, inplace=True)
df_temp.head()
```

```python
df_final = pd.concat([df_temp, temp_dis_data], ignore_index=True)
df_final.head()
```

```python
save_name = 'disease.csv'
df_final.to_csv(save_name, mode = 'w')
```
