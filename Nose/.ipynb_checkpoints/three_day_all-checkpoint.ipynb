```python
import pandas as pd
import numpy as np
```

```python
all_data = pd.read_csv('all_in_one.csv')
all_data['일교차'] = all_data['최고 기온'] - all_data['최저 기온']
# all_data['인원수'] = all_data['인원수'].round(0)
all_data['평균 기온'] = all_data['평균 기온'].round(1)
all_data['일교차'] = all_data['일교차'].round(1)
all_data['강수량'] = all_data['강수량'].round(2)
all_data['풍속'] = all_data['풍속'].round(2)
all_data['습도'] = all_data['습도'].round(2)
all_data['date'] = pd.to_datetime(all_data['date'], format='%Y-%m-%d')
all_data['day name'] = all_data['date'].dt.day_name()
all_data['강수량'].fillna(-1, inplace =True)
all_data['강수량'] = all_data['강수량'] + 1
all_data.drop(['Unnamed: 0','최고 기온', '최저 기온'],axis = 1,inplace = True)

all_data.tail(n=10)
```

```python
all_data.rename(columns={'시도':'city','인원수':'people','년':'year','월':'mon','일':'day','평균 기온':'avg temp','강수량':'rain','풍속':'wind','습도':'humid','일교차':'diff temp'}, inplace = True)
```

```python
all_data_before = all_data.groupby('city').mean()
all_data_before.head(n= 30)
```

```python
def rescale_city(df):
    if df['city'] == '서울':
        return df['people'] / 1.854383932
    elif df['city'] == '부산':
        return df['people'] / 0.645809193
    elif df['city'] == '대구':
        return df['people'] / 0.467168791
    elif df['city'] == '인천':
        return df['people'] / 0.569911163
    elif df['city'] == '광주':
        return df['people'] / 0.287369641
    elif df['city'] == '대전':
        return df['people'] / 0.289687138
    elif df['city'] == '울산':
        return df['people'] / 0.220162225
    elif df['city'] == '경기':
        return df['people'] / 2.588837389
    elif df['city'] == '강원':
        return df['people'] / 0.292584009
    elif df['city'] == '충북':
        return df['people'] / 0.315179606
    elif df['city'] == '충남':
        return df['people'] / 0.425646968
    elif df['city'] == '전북':
        return df['people'] / 0.346079567
    elif df['city'] == '전남':
        return df['people'] / 0.340672074
    elif df['city'] == '경북':
        return df['people'] / 0.512746234
    elif df['city'] == '경남':
        return df['people'] / 0.646967941
    elif df['city'] == '제주':
        return df['people'] / 0.129393588
    else:
        return None
```

```python
all_data_after = all_data.groupby('city').mean()
all_data_after.head(n= 30)
```

```python
all_data['people'] = all_data.apply(rescale_city, axis = 1)
```

```python
# 3일치 데이터 받기
temp_arr = []

all_data_group = all_data.groupby('city')

for key, group in all_data_group:
    group['SO2 -1'] = group['SO2'].shift(periods=1)
    group['SO2 -2'] = group['SO2'].shift(periods=2)
    group['SO2 -3'] = group['SO2'].shift(periods=3)
    group['O3 -1'] = group['O3'].shift(periods=1)
    group['O3 -2'] = group['O3'].shift(periods=2)
    group['O3 -3'] = group['O3'].shift(periods=3)
    group['NO2 -1'] = group['NO2'].shift(periods=1)
    group['NO2 -2'] = group['NO2'].shift(periods=2)
    group['NO2 -3'] = group['NO2'].shift(periods=3)
    group['PM10 -1'] = group['PM10'].shift(periods=1)
    group['PM10 -2'] = group['PM10'].shift(periods=2)
    group['PM10 -3'] = group['PM10'].shift(periods=3)
    group['PM25 -1'] = group['PM25'].shift(periods=1)
    group['PM25 -2'] = group['PM25'].shift(periods=2)
    group['PM25 -3'] = group['PM25'].shift(periods=3)
    group['avg temp -1'] = group['avg temp'].shift(periods=1)
    group['avg temp -2'] = group['avg temp'].shift(periods=2)
    group['avg temp -3'] = group['avg temp'].shift(periods=3)
    group['rain -1'] = group['rain'].shift(periods=1)
    group['rain -2'] = group['rain'].shift(periods=2)
    group['rain -3'] = group['rain'].shift(periods=3)
    group['wind -1'] = group['wind'].shift(periods=1)
    group['wind -2'] = group['wind'].shift(periods=2)
    group['wind -3'] = group['wind'].shift(periods=3)
    group['humid -1'] = group['humid'].shift(periods=1)
    group['humid -2'] = group['humid'].shift(periods=2)
    group['humid -3'] = group['humid'].shift(periods=3)
    group['diff temp -1'] = group['diff temp'].shift(periods=1)
    group['diff temp -2'] = group['diff temp'].shift(periods=2)
    group['diff temp -3'] = group['diff temp'].shift(periods=3)
    
    print(key)
    print(group.head())
    print('\n')
    print(group.tail())
    print('\n')
    
    temp_arr.append(group)
```

```python
all_data_three = pd.concat(temp_arr)
all_data_three.head()
```

```python
all_data_three['people'].plot(figsize=(15,3))
```
